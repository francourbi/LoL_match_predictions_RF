# Dataset analysis

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing and exploring the dataset

dataset = pd.read_csv('games.csv')

display(dataset.head())
display(dataset.info())
display(dataset.describe(include="all"))
```

The dataset has no null values, categorical variables. Feature scaling is not needed.

```python
#Selecting features

#X = dataset[['t1_towerKills','t1_inhibitorKills','t2_towerKills','t2_inhibitorKills']]
#X = dataset[['t1_baronKills', 't1_dragonKills', 't2_riftHeraldKills', 't2_baronKills', 't2_dragonKills','t2_riftHeraldKills']]
#X = dataset[['firstTower', 'firstBlood']]
#y = dataset['winner']

FEATURES = ['t1_towerKills','t1_inhibitorKills','t2_towerKills','t2_inhibitorKills']
LABEL = "winner"

X = dataset.loc[:, FEATURES].values
y = dataset.loc[:, LABEL].values
```

```python
# Splitting the dataset into the Training set and Test set

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Modeling

## Random forest

```python
from sklearn.ensemble import RandomForestClassifier
forest_classifier = RandomForestClassifier()
forest_classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = forest_classifier.predict(X_test)
```

## Randomized search

A randomized search evalutes a given number of random combinations by selecting a random value for each hyperparameter at every iteration (this is used to get the best combination of hyperparameters)

```python
#Use a Randomized Search to get the better hyperparameters for the Random Forests


from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {'n_estimators': (10, 50, 100, 1000),
               'max_features': ['auto', 'sqrt'],
               'max_depth': range(2,20),
               'min_samples_split': [2, 5, 10],
               'min_samples_leaf': [1, 2, 4],
               'bootstrap': [True, False]}

forest_classifier = RandomForestClassifier()
rnd_search = RandomizedSearchCV(forest_classifier, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error')

grid_result = rnd_search.fit(X_train, y_train)
best_params = grid_result.best_params_

print(best_params)
```

```python
# Predicting the Test set results
y_pred = grid_result.predict(X_test)
```

# Model evalutaion

```python
#Evaluating the model

from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report, confusion_matrix 

print("Mean squared error", mean_squared_error(y_test, y_pred))
print("Final RMSE", np.sqrt(mean_squared_error(y_test, y_pred)))

print("Classification report", classification_report(y_test, y_pred))
print("Confusion matrix", confusion_matrix(y_test, y_pred)) 
```
